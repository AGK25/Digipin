```python
import geopandas as gpd
import json
```

```python
with open("pincode_boundary.geojson", "r") as f:
    data = json.load(f)

gdf = gpd.GeoDataFrame.from_features(data["features"])

# Step 3: Check CRS (should be EPSG:4326)
print("Original CRS:", gdf.crs)
```

```python
# Set CRS if not set
gdf.set_crs(epsg=4326, inplace=True)
```

```python
gdf = gdf.to_crs(epsg=3857)  # meters
```

```python
# Step 5: Compute area (in square meters)
gdf["area_m2"] = gdf.geometry.area
```

```python
# Step 6: Find smallest area region
smallest_region = gdf.loc[gdf["area_m2"].idxmin()]

# Step 7: Display result
print("Smallest Area Region:")
print(smallest_region[["Pincode", "Office_Name", "area_m2"]])

# Optional: convert to km²
gdf["area_km2"] = gdf["area_m2"] / 1e6

print("\nSmallest Area (km²):", smallest_region["area_m2"] / 1e6)
```

```python
largest_region = gdf.loc[gdf["area_m2"].idxmax()]

print("Largest Area Region:")
print(largest_region[["Pincode", "Office_Name", "area_m2"]])

gdf["area_km2"] = gdf["area_m2"] / 1e6

print("\nLargest Area (km²):", largest_region["area_m2"] / 1e6)
```

```python
A_pin = largest_region[['area_m2']][0]
print(f"Largest Area: {A_pin}")

grid_resolution = 64
N = A_pin / (grid_resolution ** 2)

print(f"N = {int(N)}")

print(f"Max unique combinations = {456976}")
```
